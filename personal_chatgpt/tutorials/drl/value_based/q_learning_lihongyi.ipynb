## basics

- value-based 的方法，learned 的并不是 Policy（actor），learned 的是 Critic？
    - a critic 不直接决定 action；
    - 而是给定 actor（$\pi$，policy, under the policy），用来评估这个 actor 有多好（how good the actor is）

## $V^\pi(s)$

- 状态价值函数 state-value function $V^\pi(S)$
    - $V^\pi$ 接受 state 作为输入，输出一个 scalar 的 value，刻画累积reward；
    - 表示的是当选择使用 actor $\pi$，cumulated reward expects to be obtained after visiting state $s$
    - 它必须绑定在一个 actor（policy，$\pi$），即同一个状态 $s$，不同的 actor 视角下，其价值也会不同；
